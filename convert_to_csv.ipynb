```python
import pandas as pd
import os

# Configuration
bucket_name = 'my-pharma-data-project'
input_folder = 'canada-adverse-data/cvponline_extract_20241130/'
output_folder = 'canada-adverse-data/csv_format/'

# List of files to convert
files_to_convert = [
    'drug_product_ingredients.txt', 'drug_products.txt', 'gender_lx.txt',
    'literature_reference.txt', 'outcome_lx.txt', 'reactions.txt',
    'report_drug.txt', 'report_drug_indication.txt', 'report_links.txt',
    'report_type_lx.txt', 'reports.txt', 'seriousness_lx.txt', 'source_lx.txt'
]

print("Starting conversion process...")

for filename in files_to_convert:
    input_path = f"gs://{bucket_name}/{input_folder}{filename}"
    local_filename = filename

    print(f"\nProcessing: {filename}")

    # Download locally
    !gcloud storage cp {input_path} {local_filename}

    try:
        # Load with '$' separator (discovered in our previous study)
        df = pd.read_csv(local_filename, sep=r'\$', engine='python', encoding='latin1')

        # Save as CSV locally
        csv_filename = filename.replace('.txt', '.csv')
        df.to_csv(csv_filename, index=False)

        # Upload back to the cloud
        output_path = f"gs://{bucket_name}/{output_folder}{csv_filename}"
        !gcloud storage cp {csv_filename} {output_path}

        print(f"Successfully saved: {output_path}")

    except Exception as e:
        print(f"Failed to process {filename}: {e}")

print("\nAll files have been converted and uploaded to gs://my-pharma-data-project/canada-adverse-data/csv_format/")
```
