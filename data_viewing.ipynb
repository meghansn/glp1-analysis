```python
import pandas as pd

def load_cloud_data(bucket_path):
    # Ensure the path starts with gs://
    if not bucket_path.startswith('gs://'):
        full_path = f'gs://{bucket_path}'
    else:
        full_path = bucket_path

    # Read with tab separation (common for these TXT files)
    # We add encoding='latin1' just in case there are special characters
    return pd.read_csv(full_path, sep='\t', encoding='latin1')

```

```python
import pandas as pd

def load_canada_data(path):
    # These files appear to use '$' as a separator and have quotes around values
    return pd.read_csv(path, sep=r'\$', engine='python', encoding='latin1')

# Test it on reports.txt
df_reports = load_canada_data('gs://my-pharma-data-project/canada-adverse-data/cvponline_extract_20241130/reports.txt')
print("Fixed Reports Columns:", df_reports.columns.tolist())
display(df_reports.head(3))
```
